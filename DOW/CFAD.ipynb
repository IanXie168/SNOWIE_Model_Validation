```python
for f in file_list:
    picname = f[9:14]+"Z_NARR_CFAD.png"
    titlename = f[9:14]+"Z NARR CFAD"
    print(f)
    filename = f
    ncfile2 = Dataset(filename)
    ncfile = h5.File(filename, 'r')
    ds = netCDF4.Dataset(filename)
    zhh = ncfile2['Zhh']
    lat = ncfile['xlat']
    lon = ncfile['xlong']
    height = ncfile2['height']
    #Define x and y 0, we will loop throug heach point in the wrf array
    x = 0
    y = 0
    ct = 0
    #Create the new arrays we will put the values in for a specific height
    crsim_reflectivity = np.zeros((141, 141))
    testy = np.zeros((141, 141))

    #Create lists to add the heights and reflectivity for each level in
    total_heights = []
    total_crsim_reflectivity = []
    total_bins = 0
    #Loop through each bound:
    while ct < len(lower_bound):

        #Create a new array to fill reflectivity and heights in
        crsim_reflectivity = np.zeros((141, 141))
        testy = np.zeros((141, 141))
        print(ct)

        #Loop through each point
        while y < 141:
            while x < 141:

                #Find where the height where they are greater than the lower bound, and higher than the upper bound
                #Specified earlier 
                w = np.argwhere(((height[:,y,x] >= lower_bound[ct]) & (height[:,y,x] <= upper_bound[ct])))

                #Determine if the w list is odd or even, pick the middle index
                if len(w) > 0:
                    if len(w)%2 > 0:
                        index = w[int((len(w)-1)/2)]
                    if len(w)%2 == 0:
                        index = w[int(len(w)/2)]
                    crsim_reflectivity[y,x] = zhh[index,y,x]
                    testy[y,x] = height[index, y,x]
                    if crsim_reflectivity[y,x] > -30:
                        total_bins = total_bins + 1
                if len(w) == 0:
                    crsim_reflectivity[y,x] = np.nan
                    testy[y,x] = 0
                x = x + 1
            y = y + 1
            x = 0

        total_crsim_reflectivity.append(crsim_reflectivity)
        total_heights.append(testy)
        #Reseting x and y for the next loop
        x = 0
        y = 0
        ct = ct + 1
    #Need to count all of the bins
    x = 0
    y = 0
    z = 0

    #Make the CFAD Array
    CFAD = np.zeros((21, 61))

    tots = 0
    #Looping through each level, there are 21 of them starting from 1.0 km to 5 km
    while z < 21:
        cts = 0
        while cts < len(low):
            filler = 0

            #Finding if in a given x, y plane, if the value is within a lower or upper bound
            w=np.argwhere(((np.array(total_crsim_reflectivity)[z,:,:] >= low[cts]) & (np.array(total_crsim_reflectivity)[z,:,:] <= high[cts])))

            filler = filler + w.shape[0]
            tots = tots+w.shape[0]


            #Find the percentage of that 
            percentage = filler/total_bins

            #Add it to the CFAD
            CFAD[z,cts] = percentage
            cts = cts + 1


        z = z + 1
        
    CFAD = CFAD * 1000 #Normalize it
    fig, ax = plt.subplots()
    CS = ax.contourf(xs, ys, CFAD,50, cmap = 'jet')
    plt.title(titlename, fontsize = 15)
    plt.xlabel("dBZ", fontsize = 12)
    plt.ylabel("MSL", fontsize = 12)
    cbar = plt.colorbar(CS, ax=ax)
    cbar.ax.tick_params(labelsize=12)
    ax.tick_params(axis='both', which='major', labelsize=10)
    cbar.set_label('Normalized Frequency ($10^-1$)', fontsize = 12)
    plt.savefig(picname,dpi = 400)
```
